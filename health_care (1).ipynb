```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("healthcare_dataset.csv")
```

```python
print("Dataset Loaded Successfully!")
print("Shape:", df.shape)
```

```python
df.info()
```

```python
df.head()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.fillna("unkown", inplace=True)
```

```python
df.info()
```

Objective

The goal of this project is to analyze a healthcare dataset (from Kaggle: Prasad22 Healthcare Dataset) to identify patterns, relationships, and health risk factors among patients.

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
if 'Age' in df.columns:
  df['Age'] = df['Age'].astype(int)
print("Data cleaned successfully!")
```

```python
df.info()
```

```python
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution of Patients")
```

```python
plt.show()
```

```python
# Gender distribution (if column exists)
if 'Gender' in df.columns:
  sns.countplot(x='Gender', data=df, palette='pastel')
  plt.title("Gender Distribution")
  plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])



```

```python
# Plot heatmap again
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.to_csv("cleaned_healthcare_data.csv", index=False)
```
